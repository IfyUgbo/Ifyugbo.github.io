## Adidas Data Dictionary

- **Retailer**: The name of the retailer
- **Retailer ID**: The unique identifier of the retailer
- **Invoice Date**: The date when the invoice was generated
- **Region**: The region where the retailer operates
- **State**: The state where the retailer operates
- **City**: The city where the retailer operates
- **Product**: The name or description of the product
- **Price per Unit**: The price of a single unit of the product
- **Units Sold**: The number of units sold for the product
- **Total Sales**: The total sales amount for the product
- **Operating Profit**: The profit generated by the retailer after deducting operating expenses
- **Operating Margin**: The percentage of operating profit in relation to total sales
- **Sales Method**: The method used by the retailer for sales, such as in-store, online, etc.

# Handling Outliers
In this code, the plot size is set to (15, 8) using the plt.figure() function, ensuring that the size is applied to all box plots created. The plt.ion() function enables interactive mode for the plot, while plt.ioff() turns it off. The specified list of columns determines which columns will be plotted in the box plot. The for loop iterates over each column in the list, creating a box plot using the sns.boxplot() function. The plt.title() function adds a title to each plot, using the corresponding column name. Finally, plt.show() is used to display the plot. By running this code, separate box plots will be generated for each column in the list, with a larger plot size and an interactive plot.

```python
df2 = ['Price per Unit','Units Sold','Total Sales','Operating Profit','Operating Margin']

#df2 = ['Price per Unit','Units Sold','Total Sales','Operating Profit','Operating Margin']

# Create an interactive plot
plt.ion()

# Create a box plot for each column
for column in df2:
    sns.boxplot(x=df[column])
    plt.title(f'Box Plot of {column}')

    plt.show()
```

```python
# Convert the 'date' column to datetime64[ns] data type
df['Invoice Date'] = pd.to_datetime(df['Invoice Date'])
```

```python
# Calculate the IQR (Interquartile Range) for each column
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Identify the outliers for each column
outliers = {}
for column in df.columns:
    if column in Q1.index and column in Q3.index and column in IQR.index:
        # Drop rows with NaN values for the current column
        cleaned_column = df[column].dropna()
        
        # Check if each value in the column is an outlier
        column_outliers = ((cleaned_column < (Q1[column] - 1.5 * IQR[column])) |
                           (cleaned_column > (Q3[column] + 1.5 * IQR[column])))
        
        # Save the outliers for the current column
        outliers[column] = cleaned_column[column_outliers]
    else:
        print(f"Column '{column}' not found in quartile calculations.")

# Print the outliers for each column
for column, outlier_values in outliers.items():
    print(f"Outliers for column '{column}':")
    print(outlier_values)
    print()
```

# Creating correlation matrix with absolute values of outliers
A correlation matrix with absolute values shows the magnitude of pairwise correlations between variables in a dataset, disregarding the direction of the relationship. It is obtained by taking the absolute value of each correlation coefficient, which ranges from -1 to 1. A value of -1 represents a strong negative correlation, 0 indicates no correlation, and 1 represents a strong positive correlation. The absolute correlation matrix focuses on the strength of the relationship between variables, irrespective of its positive or negative nature.

```python
# Compute the correlation matrix
corr = df_outliers.corr()

# Take the absolute value of the correlation matrix
abs_corr = abs(corr)

# Plot the heatmap
plt.figure(figsize=(21, 17))
sns.heatmap(abs_corr, annot=True, cmap='viridis')

# Show the plot
plt.show()
```
